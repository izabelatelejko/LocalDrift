```python
# Decode byte strings to regular strings
def decode_dataframe(df):
    """Convert byte strings in dataframe to regular strings"""
    df_decoded = df.copy()
    for col in df_decoded.columns:
        if df_decoded[col].dtype == object:
            try:
                df_decoded[col] = df_decoded[col].str.decode('utf-8')
            except Exception:
                pass
    return df_decoded

# Apply decoding to all dataframes
# spam_corpus_df = decode_dataframe(spam_corpus_df)
# usenet1_df = decode_dataframe(usenet1_df)
# usenet2_df = decode_dataframe(usenet2_df)
# airlines_df = decode_dataframe(airlines_df)
poker_df = decode_dataframe(poker_df)
covtype_df = decode_dataframe(covtype_df)
```
